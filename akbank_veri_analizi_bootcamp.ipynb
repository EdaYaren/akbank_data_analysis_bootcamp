```python
def column_detection(dataframe: pd.DataFrame, cat_th: int = 10, car_th: int = 20):
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                       dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                    dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    # num_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]

    print("\n##################### Categoric #####################")
    print(cat_cols)
    print("\n##################### Numeric #####################")
    print(num_cols)
    print("\n##################### Categoric But Cardinal #####################")
    print(cat_but_car)
    print("\n##################### Numeric But Categoric #####################")
    print(num_but_cat)
    print("\n##################### Columns Overview #####################")
    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f'cat_cols: {len(cat_cols)}')
    print(f'num_cols: {len(num_cols)}')
    print(f'cat_but_car: {len(cat_but_car)}')
    print(f'num_but_cat: {len(num_but_cat)}')
    return cat_cols, num_cols, cat_but_car, num_but_cat
```

Değişken isimlerinde büyük/küçük harf uyumu sağlanmıştır.

```python
#Eksik değer tablosu
def eksik_deger_tablosu(df):
    eksik_deger = df.isnull().sum()
    eksik_deger_yuzde = 100 * df.isnull().sum()/len(df)
    eksik_deger_tablo = pd.concat([eksik_deger, eksik_deger_yuzde], axis=1)
    eksik_deger_tablo_son = eksik_deger_tablo.rename(
    columns = {0 : 'Eksik Değerler', 1 : '% Değeri'})
    return eksik_deger_tablo_son
```

```python
 eksik_deger_tablosu(dataset)
```

**7.DEĞİŞKENLER ARASI İLİŞKİ ANALİZİ**

```python
from sklearn.preprocessing import LabelEncoder

# Veri setindeki kategorik değişkenlerin adlarını alalım
categorical_columns = dataset.select_dtypes(include=['object','category']).columns

# LabelEncoder nesnesini oluşturalım
label_encoder = LabelEncoder()

# Her bir kategorik değişkeni label encoding ile dönüştürelim
for column in categorical_columns:
    dataset[column] = label_encoder.fit_transform(dataset[column])

dataset.describe()
```

```python
# Belirli bir eşik değeri belirleyelim
threshold = 0.5

# Korelasyon matrisindeki sadece eşik değerinden büyük korelasyon katsayılarını alalım
# Korelasyon matrisini oluşturalım
corr_matrix = dataset.corr().abs()
high_corr = corr_matrix[corr_matrix > threshold]

# Isı haritasını çizelim
plt.figure(figsize=(12, 8))
sns.heatmap(high_corr, annot=True, cmap='GnBu', linewidths=0.2, vmin=0, vmax=1)
plt.xlabel('Özellikler')
plt.ylabel('Özellikler')
plt.title('Korelasyon Matrisi (Eşik Değerinden Büyük Korelasyonlar)')
plt.show()
```

Kategorik olan veriler label encoder ile sayısal veriye dönüştürülmüştür. Değişkenler arasındaki ilişki 0.5 eşik değeri olarak belirlenip eşik değerinden büyük ilişkisi olan değişkenler ekrana yansıtılmıştır. Korelasyon grafiği incelendiğinde en yüksek ilişkiye uçağın dakika cinsinden varış ve kalkış gecikmesi olduğunu ve aralarında %96'lık bir ilişki olduğu görülmektedir.

**8.FEATURE SELECTION**

LASSO (En Az Mutlak Büzülme ve Seçim Operatörü), aşırı uyumu önlemek için hem özellik seçimini hem de düzenlemeyi gerçekleştiren doğrusal regresyon yönteminin bir modifikasyonudur. LASSO, daha az önemli özelliklerin regresyon katsayılarını sıfıra indirerek bunları modelden etkili bir şekilde çıkarır. Geriye kalan sıfır olmayan katsayılar önemli özellikleri göstermektedir.

```python
from sklearn.linear_model import LassoCV

# Hedef değişkeni tanımlayın
target = dataset['Satisfaction']

# Veri setinden hedef değişkeni çıkarın
X = dataset.drop(columns=['Satisfaction'])

# LASSO regresyon modelini oluşturalım
lasso = LassoCV(cv=5, random_state=0)
lasso.fit(X, target)

# Katsayıları çizelim
plt.figure(figsize=(10, 6))
plt.plot(range(len(X.columns)), lasso.coef_, marker='o', markersize=8, linestyle='None')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=2)
plt.xticks(range(len(X.columns)), X.columns, rotation=90)
plt.ylabel('Katsayılar')
plt.xlabel('Özellikler')
plt.title('LASSO Regresyonunu Kullanarak Özelliğin Önemi')
plt.show()
```

Elde edilen grafik incelendiğinde;
* Uçak içi eğlence değişkeninin müşteri memnuniyeti üzerinde en fazla etkiye sahip olduğu,
* Kalkış/Varış saati uygunluğu değişkeninin ise müşteri memnuniyeti üzerinde en az etkiye sahip olduğu

gözlemlenmiştir.

Lasso regresyon ile müşteri memnuniyeti üzerinde etkisi olmayan değişkenlerin ise katsayısı "0" olarak atanmıştır. Bu değişkenler;
